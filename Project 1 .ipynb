# To analyze sales trends using the techniques you mentioned, we can utilize Python libraries such as pandas for data manipulation and matplotlib or seaborn for data visualization. Here's how you can perform each analysis:

## Month-wise Sales Trend:

```python
monthly_sales = data.groupby(['Year', 'Month'])['Total Revenue'].sum().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(x='Month', height='Total Revenue', data=monthly_sales,color='c')
sns.lineplot(x='Month', y='Total Revenue', data=monthly_sales, marker='o', color='red')

plt.title('Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)

plt.show()
```

* Highly sales Revenue between 11,2 and 4 Month of the year 

## Year-Wise Trend

```python


# Plotting

plt.figure(figsize=(12, 6))
plt.bar(x='Year', height='Total Revenue', data=monthly_sales,color='c')
sns.lineplot(x='Year', y='Total Revenue', data=monthly_sales, marker='o', color='red')
plt.title('Year-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)

plt.show()
```

```python

# Pivot the data to have years as columns and months as rows
yearly_monthly_sales = data.pivot_table(index='Month', columns='Year', values='Total Revenue', aggfunc='sum')

# Plotting
plt.figure(figsize=(12, 6))
sns.heatmap(yearly_monthly_sales, cmap='YlGnBu', annot=True, fmt='.0f', linewidths=.5)
plt.title('Yearly Month-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Month')
plt.tight_layout()
plt.show()
data.to_csv("yearly_monthly_sales.csv",index=False)
```

## "Yearly Monthly" analysis provides valuable insights into the seasonality and trends of sales behavior, empowering businesses to make data-driven decisions and optimize their operations for improved performance and profitability throughout the year.
